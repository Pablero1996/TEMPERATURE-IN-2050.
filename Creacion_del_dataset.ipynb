## CREACIÓN DEL DATASET


Lectura de datos

```python
#Leemos el primer dataset CAMBIO.CLIMATICO
import pandas as pd
csv_path = 'DATOS.CAMBIO.CLIMATICO.csv'
df_cambio_climatico=pd.read_csv(csv_path, sep=";")
df_cambio_climatico
```

```python
df_cambio_climatico.isna().sum()
```

```python
df_cambio_climatico["Country or Area"].unique()
```

```python
#Quitamos la columna "Nf3" que no nos aporta gran información.
df_cambio_climatico=df_cambio_climatico.drop(df_cambio_climatico.columns[[6]],axis=1)
df_cambio_climatico
```

```python
#Leemos el segundo datadet ENERGÍAS RENOVABLES.

csv_path="modern-renewable-energy-consumption.csv"
renewable=pd.read_csv(csv_path, sep=",")
renewable
```

```python
renewable.isna().sum()
```

```python
#Cambiamos el nombre de algunas variables que nos facilitaraá las cosas.
headers=['Country or Area', 'Code', 'Year', 'Hydropower (terawatt-hours)',
       'Solar (terawatt-hours)', 'Wind (terawatt-hours)',
       'Other renewables (terawatt-hours)']
renewable.columns=headers

renewable["Country or Area"] = renewable["Country or Area"].replace("United States", "United States of America")

renewable["Country or Area"] = renewable["Country or Area"].replace("Russia", "Russian Federation")

renewable["Country or Area"] = renewable["Country or Area"].replace("Czech Republic", "Czechia")



renewable["Country or Area"].unique()
```

```python
#Leemos el tercer dataset TEMPERATURAS.

csv_path="TEMPERATURAS.csv"
temperaturas=pd.read_csv(csv_path,error_bad_lines=False,sep=";")
temperaturas
```

```python
temperaturas.isna().sum()
```

```python
temperaturas["Area"] = temperaturas["Area"].replace("United Kingdom of Great Britain and Northern Ireland", "United Kingdom")

temperaturas.dtypes
```

```python
temperaturas1=temperaturas[["Area","Year", "Value"]]
headers=["Country or Area","Year","Temperature"]
temperaturas1.columns = headers
temperaturas1
```

```python
cruze1=pd.merge(df_cambio_climatico,temperaturas1,on=['Country or Area','Year'])
cruze1
#cruze1.isna().sum()
#cruze1["Country or Area"].unique()
```

```python
cruze2=pd.merge(cruze1,renewable,on=['Country or Area','Year'])
cruze2
#cruze2["Country or Area"].unique()
```

```python
cruze2.to_csv('datasetCruzados.csv')
```
